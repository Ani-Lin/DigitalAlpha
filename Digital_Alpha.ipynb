```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
home_values_df['2022-05-31'].fillna(value=home_values_df['2022-05-31'].mean(), inplace=True)
home_values_df['2022-06-30'].fillna(value=home_values_df['2022-06-30'].mean(), inplace=True)
home_values_df['2022-07-31'].fillna(value=home_values_df['2022-07-31'].mean(), inplace=True)
home_values_df['2022-08-31'].fillna(value=home_values_df['2022-08-31'].mean(), inplace=True)
home_values_df['2022-09-30'].fillna(value=home_values_df['2022-09-30'].mean(), inplace=True)
```

```python
home_values_df.isnull().sum()
```

```python
import matplotlib.pyplot as plt

test_df=home_values_df.groupby(['RegionName'])['2022-05-31'].max()
test_df
```

```python
#fig = plt.figure(figsize = (20, 20))
 
# creating the bar plot]

test_df.plot(kind='line', title='Region wise max price', ylabel='Price',xlabel='Region', figsize=(20, 10))
plt.show()
```
